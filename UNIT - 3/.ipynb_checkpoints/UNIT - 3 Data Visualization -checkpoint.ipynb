```python
# ticks are the markers denoting data points on axes
# importing libraries


#values of x and y axes
x = [5,10,15,20,25,30,35,40,45,50]
y = [1,4,3,2,7,6,9,8,10,5]

plt.plot(x,y,'g')
plt.xlabel('x')    
plt.ylabel('y')
         
         
# here we set the size for ticks, rotation and color value
plt.tick_params(axis='x',labelsize=18,labelrotation=90, labelcolor='r')
plt.tick_params(axis='y',labelsize=18,labelrotation=20, labelcolor='r')     
         
         
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np

#draw plt as fig


fig, ax = plt.subplots()

x = np.linspace(0,20,1000)
ax.plot(x, np.cos(x))
ax.axis('equal')


#text annotation to the plot where it indicate maximum value of the 

ax.annotate('local maximum', xy=(6.28,1), xytext=(10,4),
           arrowprops=dict(facecolor='black', shrink=0.04))
#text annotation to the plot where it indicater minimum value of the
ax.annotate('local minimum',xy=(5* np.pi, -1), xytext=(2,-6),
            arrowprops=dict(arrowstyle="->",
                           connectionstyle="angle3,angleA=0,angleB=-90"));
```

# adding annotate and legend

```python
import numpy as np
import matplotlib.pyplot as plt
#you can also write 
#from matplotlib import pyplot as plt



x = np.arange(0,10)
y = x^2
z = x^3


plt.plot(x,y)
plt.plot(y,z)
plt.plot(x,z)

plt.title('Graph drawing')
plt.xlabel('Time')
plt.ylabel('Distance')




#annotation

plt.annotate(xy=[2,1], text="First Entry", rotation = 35)
plt.annotate(xy=[6,5], text="Second Entry" , rotation = 35)
plt.annotate(xy=[8,10], text="Third Entry" , rotation = 35)

#legend

plt.legend(['blue' , 'orange', ' green'], loc = 2)

plt.show()
```

# Pie Chart

pie chart is a circular statical plot that can display only one series of data.

the area of the chart is the total % of the given data

```python
#import libreries

#creating dotset

Branch = ["Computer", "Electrical", "Mechanical" , "Civil", "Automobile", "IT"]

data = [50, 34, 40, 29, 12, 23]
explode=(0.15,0,0,0,0.1,0)

#creat chart

fig = plt.figure(figsize=(10,9))
plt.pie(data, labels= Branch, explode=explode, autopct = '%1.1f%%' , shadow = True)

plt.show()
```

# 3D ploting

```python
#import mplot3d toolkit, numopy & matplotlib

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d #3d toolkit lib
```

```python
fig = plt.figure()

#syntex for 3D projecxtion

ax = plt.axes(projection = '3d')

#defining all 3 axes


z= np.linspace(0,1,100)
x = z * np.sin(25*z)
y = z * np.cos(25*z)

#3d ploting

ax.plot3D(x, y, z, 'green')
ax.set_title('3D ploting')

plt.show()
```
